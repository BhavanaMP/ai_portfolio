An image is nothing but a NumPy array or a matrix.

An array is a vector with one or more dimensions. A one-dimensional array can be considered a vector, and an array with two dimensions can be considered a matrix

While the term "matrix" often implies a 2D array, the concept can be extended to higher dimensions. These are then referred to as tensors. A 3D tensor would have three indices, representing rows, columns, and a third dimension. Think of it as a stack of matrices.

```python
from skimage import data, io
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px

datasets = (
    'hubble_deep_field',
    'immunohistochemistry',
    'lily',
    'microaneurysms',
    'moon',
    'retina',
    'shepp_logan_phantom',
    'skin',
    'cell',
    'human_mitosis',
)

for name in datasets:
    caller = getattr(data, name)
    img = caller()
    plt.figure()
    plt.title(name)
    if img.ndim == 2:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(img)
plt.show()
    
```
