```python
#define augmentation transforms

#For ADE20K
MEAN = np.array([123.675, 116.280, 103.530]) / 255
STD = np.array([58.395, 57.120, 57.375]) / 255

def train_transforms():
    return A.Compose([
        A.RandomSizedCrop(min_max_height=(800, 800),
                            height=1024, width=1024, p=0.5),
        A.OneOf([
            A.HueSaturationValue(hue_shift_limit=0.2, sat_shift_limit=0.2,
                                 val_shift_limit=0.2, p=0.9),
            A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2,
                                       p=0.9)
        ]),
        A.ToGray(p=0.01),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.Resize(height=512, weight=512, p=1),
        A.Cutout(num_holes=8, max_h_size=64, max_w_size=64, fill_value=0,
                 p=0.5),
        A.Normalize(mean=MEAN, std=STD),
        ToTensorV2(p=1.0)
    ], p=1.0)


def val_transforms():
    return A.Compose([
        A.Resize(height=512, width=512, p=1.0),
        ToTensorV2(p=1.0),
        ],
        p=1.0)

def test_transforms():
    pass
```

```python
#Define our custom dataset
class CustomData(Dataset):
    def __init__(self, dataset, processor, transforms=None, task="panoptic"):
        self.dataset = dataset
        self.processor = processor
        self.transforms = transforms
        self.task = task

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, index):
        image = self.dataset[index]["image"]
        # panoptic_seg_mask = self.dataset[index]["label"]
        panoptic_seg_mask = self.dataset[index]["mask"]

        # apply transforms
        if self.transforms is not None:
            transformed = self.transforms(image=np.array(image), mask=np.array(panoptic_seg_mask))
            image, panoptic_seg_mask = transformed["image"], transformed["mask"]
            # convert to numpy C, H, W from albumentations H, W, C
            image = image.transpose(2, 0, 1)

        panoptic_seg_mask = rgb_to_id(panoptic_seg_mask)
        inst2class = {segment["id"]: segment["category_id"] for segment in self.dataset[index]["segments_info"]}

        #apply processor
        inputs = self.processor(images=[image], segmentation_maps=[panoptic_seg_mask],
                       instance_id_to_semantic_id=inst2class,
                       task_inputs=[self.task],
                       return_tensors="pt")

        inputs = {
            k: v.squeeze() if isinstance(v, torch.Tensor) else v[0]
            for k, v in inputs.items()
        }

```

```python
#Load the dataset and Instantiate the datasets

dataset = load_dataset("nateraw/pascal-voc-2012")
checkpoint = "shi-labs/oneformer_ade20k_swin_tiny"
processor = AutoProcessor.from_pretrained(checkpoint)

# create PyTorch datasets
train_dataset = CustomData(
    dataset=dataset["train"],
    processor=processor,
    transforms=train_transforms,
    task="panoptic"
)

val_dataset = CustomData(
    dataset=dataset["val"],
    processor=processor,
    transforms=val_transforms,
    task="panoptic"
)

test_dataset=None
```

```python
class AverageMeter(object):
    """Class for averaging loss, metric, etc., over epochs.
    Computes and stores the average and current value"""
    def __init__(self):
        self.reset()

    def reset(self):
        self.val = 0
        self.avg = 0
        self.sum = 0
        self.count = 0

    def update(self, val, n=1):
        self.val = val
        self.sum += val * n
        self.count += n
        self.avg = self.sum / self.count
```

```python
#Training Config
class  TrainGlobalConfig:
    batch_size = 2
    num_workers = 2
    epochs = 3
    lr = 5e-5
    optimizer = torch.optim.Adam
    save_every = 20 #save every 20 epochs

    step_scheduler = False # do scheduler.step after optimizer.step
    validation_scheduler = True # do scheduler.step after validation stage loss

    SchedulerClass = torch.optim.lr_scheduler.ReduceLROnPlateau
    scheduler_params = dict(
        mode="min",
        factor=0.5,
        patience=1,
        vebose=False,
        threshold=0.0001,
        threshold_mode="abs",
        cooldown=0,
        min_lr=1e-8,
        eps=1e-08
    )
    root_dir = Path("./")
    verbose = True
    verbose_step = 1
```

```python
#Training Config and Model
import warnings
warnings.filterwarnings("ignore")

import logging
        # logging.getLogger(__name__).info(
#             "[OneFormerTrainer] Full TransformGens used in training: {}".format(
#                 str(self.tfm_gens)
#             )

class OneFormerTrainer:
    def __init__(self, model, config, train_dataloader=None,
                 val_dataloader=None, device=None):
        self.model = model
        self.config = config
        self.train_dataloader = train_dataloader,
        self.val_dataloader = val_dataloader,
        self.device = device
        self.optimizer = self.config.optimizer(params=self.model.parameters(),
                                          lr=self.config.lr)
        self.scheduler = self.config.SchedulerClass(self.optimizer,
                                               **self.config.scheduler_params)
        self.start_epoch = 0
        self.best_loss = 10**5
        self.root_dir = self.config.root_dir
        self.base_dir = self.root_dir / "oneformer_FT"
        self.log_save_path = f"{self.base_dir}/log.txt"
        self.save_every = TrainGlobalConfig.save_every
        if not self.base_dir.is_dir():
            print(f"Save Directory doesn't exist.. Creating one!")
            self.base_dir.mkdir(exist=True, parents_ok=True)
        self.log(f"Trainer is prepared. Device is {self.device}")
        param_optimizer = list(self.model.named_parameters())
        no_decay = ["bias", "LayerNorm.bias", "LayerNorm.weight"]
        optimizer_grouped_parameters = [
            {
                "params": [param for name, param in param_optimizer
                           if not any(nd in name for nd in no_decay)],
                "weight_decay": 0.001
            },
            {
                "params": [param for name, param in param_optimizer
                        if any(nd in n for nd in no_decay)],
                "weight_decay": 0.0
            }
        ]
        # Define optimizer using the grouped parameters
        #self.optimizer = self.config.optimizer(
        #   params=optimizer_grouped_parameters, lr=self.config.lr)


    def save(self, save_path):
        checkpoint = {
            "model_state_dict": self.model.state_dict(),
            "optimizer_state_dict": self.optimizer.state_dict(),
            "scheduler_state_dict": self.scheduler.state_dict(),
            "best_loss": self.best_loss,
            "epoch": self.start_epoch
        }
        torch.save(obj=checkpoint, f=save_path)

    def load(self, save_path):
        checkpoint = torch.load(f=save_path)
        self.model.load_state_dict(checkpoint["model_state_dict"])
        self.optimizer.load_state_dict(checkpoint["optimizer_state_dict"])
        self.scheduler.load_state_dict(checkpoint["scheduler_state_dict"])
        self.best_loss = checkpoint["best_loss"]
        self.start_epoch = checkpoint["epoch"] + 1

    def log(self, log_message):
        if self.config.verbose:
            print(log_message)
        with open(self.log_save_path, "a+") as logger:
            logger.write(f"{log_message}\n")

    def train_step(self, model, train_dataloader, optimizer, sheduler, device):
        model.to(device)
        # num_steps = self.config.epochs * len(train_dataloader)
        summary_loss = AverageMeter()
        t = time.time()
        for idx, batch in enumerate(train_dataloader):
            model.train()
            if (self.config.verbose) and (idx % self.config.verbose_step == 0):
                print(
                    f"Train Step: {idx} / {len(train_dataloader)} | "
                    f"summary_loss: {summary_loss.avg:.5f} | "
                    f"time: {(time.time() - t):.5f} ",
                    end="\r")
            optimizer.zero_grad()
            outputs = model(
                pixel_values=batch["pixel_values"].to(device),
                mask_labels=[labels.to(device) for labels in batch["mask_labels"]],
                class_labels=[labels.to(device) for labels in batch["class_labels"]]
            )
            loss = outputs.loss
            loss.backward()
            batch_size = batch["pixel_values"].size(0)
            summary_loss.update(loss.detach().item(), n=batch_size)
            optimizer.step()
            if self.config.step_scheduler:
                self.scheduler.step()
            del batch
        return summary_loss

    def val_step(self, model, val_dataloader, device):
        model.to(device)
        model.eval()
        summary_loss = AverageMeter()
        t = time.time()
        for idx, batch in enumerate(val_dataloader):
            if (self.config.verbose) and (idx % self.config.verbose_step == 0):
                    print(
                        f"Val Step {idx} / {len(val_dataloader)}, " + \
                        f"summary_loss: {summary_loss.avg:.5f}, " + \
                        f"time: {(time.time() - t):.5f}", end="\r"
                    )
            with torch.inference_mode():
                outputs = model(
                    pixel_values=batch["pixel_values"].to(device),
                    mask_labels=[labels.to(device) for labels in batch["mask_labels"]],
                    class_labels=[labels.to(device) for labels in batch["class_labels"]]
                )
                val_loss = outputs.loss
                batch_size = batch["pixel_values"].size(0)
                summary_loss.update(val_loss.detach().item(), batch_size)
            del batch
        return summary_loss

    def train(self):
        if self.config.verbose:
                start_time = timer()
                print(f"Training started at: {start_time} ms | Epoch: {self.start_epoch}")
                lr = self.optimizer.param_groups[0]["lr"]
                timestamp = datetime.utcnow().isoformat()
                self.log(f"\n{timestamp}\nLR: {lr}")

        for epoch in range(tqdm(self.start_epoch, self.config.epochs)):
            #check if the checkpoint already exists, load from checkpoint to resume
            #.......
            #Train step
            t = time.time()
            train_summary_loss = self.train_step(
                self.model, self.train_dataloader, self.optimizer,
                self.scheduler, self.config, self.device)
            #Log and save checkpoint every epoch
            self.log(f"[RESULT]: Train. Epoch: {self.epoch} | summary_loss: {train_summary_loss.avg:.5f} | time: {(time.time() - t):.5f}")
            self.save(f"{self.base_dir}/last-checkpoint.pth")
            print(f"Epoch {epoch} | Training checkpoint saved at {self.base_dir}/last-checkpoint.pth")


            #Val step
            t = time.time()
            summary_loss = self.val_step(self.model, self.val_dataloader, self.device)
            self.log(f'[RESULT]: Val. Epoch: {self.epoch} | summary_loss: {summary_loss.avg:.5f} |time: {(time.time() - t):.5f}')

            if summary_loss.avg < self.best_loss:
                self.best_loss = summary_loss.avg
                #save the checkpoint
                if self.save_every % epoch == 0:
                     #remove the previous last 3 best checkpoints
                    for path in sorted(glob(f"{self.base_dir}/best-checkpoint-*epoch.pth"))[:-3]:
                        os.remove(path)
                    self.save(f"{self.base_dir}/best-checkpoint-{str(self.start_epoch).zfill(3)}epoch.pth")

            if self.config.validation_scheduler:
                self.scheduler.step(metrics=summary_loss.avg)

            self.start_epoch += 1
```

Some notes on setting up optimizer param groups

This code for optimizer_grouped_parameters sets up two parameter groups for the optimizer: one group with weight decay for most parameters and another group without weight decay for bias terms and certain normalization layer parameters. This allows for different optimization settings for different types of parameters in the neural network.







```python
#Dataloader and Trainer
from huggingface_hub import hf_hub_download
import json

def collate_fn(batch):
    """
    Define custom collate function which defines how to batch examples together
    This function basically handles batches to stack when they have different dimensions
    and could nt stack. So we overide the default collatefn and stack each keys in
    a batch separately
    """
    batch_dict = {}
    pixel_values = torch.stack([example["pixel_values"] for example in batch])
    pixel_mask = torch.stack([example["pixel_mask"] for example in batch])
    class_labels = [example["class_labels"] for example in batch]
    mask_labels = [example["mask_labels"] for example in batch]
    batch_dict.update({"pixel_values": pixel_values,
                       "pixel_mask": pixel_mask,
                       "class_labels": class_labels,
                       "mask_labels": mask_labels})
    return batch_dict

def get_id2label():
    repo_id = f"huggingface/label-files"
    filename = "ade20k-id2label.json"
    with open(hf_hub_download(
        repo_id=repo_id, filename=filename, repo_type="dataset"
        ), "r") as f:
        id2label = json.load(f)
        id2label = {int(k):v for k,v in id2label.items()}
        return id2label

def get_model():
    checkpoint = "shi-labs/oneformer_ade20k_swin_tiny"
    model = AutoModelForUniversalSegmentation.from_pretrained(
        checkpoint,
        id2label=get_id2label(),
        ignore_mismatched_sizes=True)
    return model

def train():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    model = get_model()
    model.to(device)

    optimizer = torch.optim.Adam(
        params=model.parameters(),
        lr=TrainGlobalConfig.lr
    )
    train_dataloader = DataLoader(
        dataset=train_dataset,
        batch_size=TrainGlobalConfig.batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        pin_memory=True,
        num_workers=TrainGlobalConfig.num_workers
    )
    val_dataloader = DataLoader(
        dataset=val_dataset,
        batch_size=TrainGlobalConfig.batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        pin_memory=False,
        collate_fn=collate_fn
    )

    trainer = OneFormerTrainer(model=model,
                               config=TrainGlobalConfig,
                               train_dataloader=train_dataloader,
                               val_dataloader=val_dataloader,
                               device=device)
    trainer.train()
```
