```python
import plotly.graph_objects as go
from datasets import load_dataset
import numpy as np

# Load the Cityscapes dataset
cityscapes_semantic = load_dataset("BhavanaMalla/cityscapes-semantic")

# Define the color map and class names
class_names = [
    "Road", "Sidewalk", "Building", "Wall", "Fence", "Pole", "Traffic light", "Traffic sign",
    "Vegetation", "Terrain", "Sky", "Person", "Rider", "Car", "Truck", "Bus", "Train", "Motorcycle", "Bicycle"
]
palette = np.array([
    (128, 64, 128),  # Road
    (244, 35, 232),  # Sidewalk
    (70, 70, 70),    # Building
    (102, 102, 156), # Wall
    (190, 153, 153), # Fence
    (153, 153, 153), # Pole
    (250, 170, 30),  # Traffic light
    (220, 220, 0),   # Traffic sign
    (107, 142, 35),  # Vegetation
    (152, 251, 152), # Terrain
    (70, 130, 180),  # Sky
    (220, 20, 60),   # Person
    (255, 0, 0),     # Rider
    (0, 0, 142),     # Car
    (0, 0, 70),      # Truck
    (0, 60, 100),    # Bus
    (0, 80, 100),    # Train
    (0, 0, 230),     # Motorcycle
    (119, 11, 32)    # Bicycle
]) / 255.0

# Extract the pixel distribution
pixel_counts = {}

for sample in cityscapes_semantic["train"]:
    mask = sample["annotation_labeltrainids"]  # Access the segmentation mask
    unique, counts = np.unique(mask, return_counts=True)
    for cls, count in zip(unique, counts):
        if cls in pixel_counts:
            pixel_counts[cls] += count
        else:
            pixel_counts[cls] = count

# Prepare data for plotting
classes = [class_names[cls] for cls in pixel_counts.keys() if cls < len(class_names)]
pixels = [pixel_counts[cls] for cls in pixel_counts.keys() if cls < len(class_names)]

# Convert pixel counts to powers of 10
pixels_log = [np.log10(p) for p in pixels]


# Map colors to classes based on the pixel count keys
colors = []
for key in pixel_counts.keys():
    if key == 255:
        # Assign black color for key 255
        colors.append(f'rgb(0, 0, 0)')
    elif key < len(class_names):
        # Map the key to the corresponding color in the palette
        r, g, b = palette[key]
        colors.append(f'rgb({r*255}, {g*255}, {b*255})')



```

```python
import torch
from collections import Counter
from datasets import load_dataset
import numpy as np
from torchvision.transforms.functional import pil_to_tensor

def compute_class_frequencies(hf_dataset, num_classes):
    """
    Compute per-class pixel frequency for a semantic segmentation test set.

    Args:
        num_classes (int): Total number of classes.

    Returns:
        class_frequencies (torch.Tensor): A tensor of shape (num_classes,) where each value
                                          represents the frequency of the corresponding class.
    """
    class_counts = torch.zeros(num_classes, dtype=torch.long)
    total_pixels = 0

    for sample in hf_dataset["train"]:
        mask = sample["annotation_labeltrainids"]  # Access the segmentation mask
        targets = pil_to_tensor(mask)
        # print(targets.shape) # Shape: (B, H, W),  torch.Size([1, 1024, 2048])

        # Deal background class
        background_mask = targets == 255
        targets[background_mask] = -1 # Set background pixels to -1 (or any ignored label value)
        # Filter out the background pixels
        targets = targets[targets != -1]  # Only keep non-background pixels
        targets = targets.view(-1) # Flatten to (N,)

        # Count pixel occurrences for each class
        class_counter = Counter(targets.tolist())
        for class_idx, count in class_counter.items():
            class_counts[class_idx] += count

        # Update total pixel count
        total_pixels += targets.numel()

    # Compute frequency as the fraction of pixels belonging to each class
    class_frequencies = class_counts.float() / total_pixels
    return class_frequencies


# Load the Cityscapes dataset
cityscapes_semantic = load_dataset("BhavanaMalla/cityscapes-semantic")

# Define the color map and class names
class_names = ["Road", "Sidewalk", "Building", "Wall", "Fence", "Pole", "Traffic light", "Traffic sign", "Vegetation", "Terrain", "Sky", "Person", "Rider", "Car", "Truck", "Bus", "Train", "Motorcycle", "Bicycle"]

num_classes = 19
class_frequencies = compute_class_frequencies(cityscapes_semantic, num_classes)

print("Train Per-Class Frequencies:", class_frequencies.tolist())
```

```python
import plotly.graph_objects as go
from plotly.io import write_image
import pickle

result_files = ["roc_result_0_Baseline.pkl", "roc_result_0_EDL.pkl", "roc_result_0_MCD.pkl",
#  "roc_result_1_OHEM.pkl",
                # "roc_result_0_MCD+OHEM.pkl", "roc_result_1_EDL+UNC.pkl", "roc_result_2_EDL+OHEM.pkl"
                ]

# Combine results into one figure
fig = go.Figure()
for result_file in result_files:
    # print(f"[INFO]: Opening the result file {result_file}")
    with open(result_file, 'rb') as f:
        print(f"[INFO]: Opening the result file {result_file}")
        roc_curve_data, auroc, fpr_at_recall_level, label = pickle.load(f)

        fpr, tpr = roc_curve_data[0], roc_curve_data[1]
        fig.add_trace(
            go.Scatter(
                x=fpr,
                y=tpr,
                mode='lines',
                name=f'{label}: AUROC={auroc:.3f}, FPR@95={fpr_at_recall_level:.3f}',
                showlegend=True,
            )
        )
        fig.add_trace(
            go.Scatter(
                x=[fpr_at_recall_level],
                y=[0.95],
                mode='markers',
                marker=dict(color='red', size=5),
                showlegend=False,
            )
        )

        # Update layout
        fig.update_layout(
            title="Combined ROC Curves for Multiple Datasets",
            xaxis=dict(
                title="False Positive Rate",
                showline=True, linewidth=1, linecolor="black", mirror=True,
                showgrid=True, gridcolor="lightgray",
            ),
            yaxis=dict(
                title="True Positive Rate",
                showline=True, linewidth=1, linecolor="black", mirror=True,
                showgrid=True, gridcolor="lightgray",
                title_font=dict(size=18),
                tickfont=dict(size=14),
                range=[0.8, 1]
            ),
            showlegend=True,
            legend=dict(
                title="<b>Legend</b>",
                x=0.99, y=0.01,
                xanchor="right", yanchor="bottom",
                bordercolor="gray",
                borderwidth=1,
                font=dict(size=12)
            ),
            plot_bgcolor="white",
            paper_bgcolor="white"
        )

fig.show()
```
