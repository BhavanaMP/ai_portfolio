```python
import pandas as pd
import plotly.express as px

# Load the CSV file
df = pd.read_csv('/content/per_image_test_results_FPN_MIT_B3_MCD_OHEM_BASELINE_FINETUNE_cityscapes_7dpnjq2.csv')

# Create a long-form DataFrame for Plotly
df_long = pd.melt(df, id_vars=['test_img_names'], value_vars=['test_nlls'], var_name='Metric', value_name='Value')

# Create the box plot with hover information
fig = px.violin(df_long, x='Metric', y='Value', points='all', hover_data=['test_img_names'])

# Customize the layout
fig.update_layout(
    title="MCD_OHEM test_nlls per Test Image",
    yaxis_title="NLL Values",
    xaxis_title=""
)

# Show the plot
fig.show()
```

```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/per_image_test_results_FPN_MIT_B3_MCD_OHEM_BASELINE_FINETUNE_cityscapes_7dpnjq2.csv')

# Calculate the standard deviation for IoU and accuracy
iou_std = df["test_IoUs"].std()
miou = df["test_IoUs"].mean()
accuracy_std = df["test_accs"].std()
macc = df["test_accs"].mean()
nll_std = df["test_nlls"].std()
mnll = df["test_nlls"].mean()
ece_std = df["test_eces"].std()
mece = df["test_eces"].mean()


# Print the results
print(f'mIoU: {miou:.4f}  {iou_std:.4f}')
print(f'mAcc: {macc:.4f}  {accuracy_std:.4f}')
print(f'mNLL: {mnll:.4f}  {nll_std:.4f}')
print(f'mECE: {mece:.4f}  {ece_std:.4f}')
```
