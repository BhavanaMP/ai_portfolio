Normality Test

```python
import pandas as pd
from scipy.stats import shapiro

# Load the CSV data into a pandas DataFrame
file_path = 'per_image_test_results_FPN_MIT_B3_baseline_cityscapes_t7z5h2i.csv'
data = pd.read_csv(file_path)

# Function to perform the Shapiro-Wilk test and return results
def shapiro_test(column):
    stat, p_value = shapiro(column)
    return {'statistic': stat, 'p_value': p_value}

# Columns to test for normality
metrics = ['test_IoUs', 'test_accs', 'test_eces']

# Apply Shapiro-Wilk test to each column and store results
results = {}
for metric in metrics:
    results[metric] = shapiro_test(data[metric])

# Print results
alpha = 0.05
for metric, result in results.items():
    print(f"{metric}: Statistic={result['statistic']}, p-value={result['p_value']}")
    if result['p_value'] < alpha:
        print(f"{metric}: Data is not normally distributed")
    else:
        print(f"{metric}: Data is normally distributed")
```

Wilcoxin Test

```python
import pandas as pd
from scipy.stats import wilcoxon

# Load the data
baseline_df = pd.read_csv('per_image_test_results_FPN_MIT_B3_baseline_cityscapes_t7z5h2i.csv')
improve_df = pd.read_csv('per_image_test_results_FPN_MIT_B3_EDL_cityscapes_rdbyxtv.csv')

# Extract the relevant metrics
# Assuming both CSVs have the same columns for metrics like 'test_IoUs', 'test_accs', etc.
baseline_IoUs = baseline_df['test_IoUs']
improve_IoUs = improve_df['test_IoUs']

baseline_accs = baseline_df['test_accs']
improve_accs = improve_df['test_accs']

baseline_eces = baseline_df['test_eces']
improve_eces = improve_df['test_eces']

# Perform the Wilcoxon signed-rank test for each metric
stats_iou, p_value_iou = wilcoxon(baseline_IoUs, improve_IoUs)
stats_accs, p_value_accs = wilcoxon(baseline_accs, improve_accs)
stats_eces, p_value_eces = wilcoxon(baseline_eces, improve_eces)

# Output the results
print("Wilcoxon Test Results:")

print(f"IOUs: Test Statistic = {stats_iou}, p-value = {p_value_iou}")
if p_value_iou < 0.05:
    print("Statistically significant difference in IOUs")
else:
    print("No statistically significant difference in IOUs")

print(f"Accuracy: Test Statistic = {stats_accs}, p-value = {p_value_accs}")
if p_value_accs < 0.05:
    print("Statistically significant difference in Accuracy")
else:
    print("No statistically significant difference in Accuracy")

print(f"ECEs: Test Statistic = {stats_eces}, p-value = {p_value_eces}")
if p_value_eces < 0.05:
    print("Statistically significant difference in ECEs")
else:
    print("No statistically significant difference in ECEs")
```

STatistical Test

```python
import pandas as pd
import scipy.stats as stats

# Load CSV files (Assuming each CSV has columns: 'IoU', 'ACC', 'ECE')
baseline_df = pd.read_csv("per_image_test_results_baseline.csv")
mcd_df = pd.read_csv("per_image_test_results_MCD.csv")
ohem_df = pd.read_csv("per_image_test_results_Baseline_OHEM.csv")
edl_df = pd.read_csv("per_image_test_results_EDL.csv")

# Function to compute median and IQR
def compute_median_iqr(df, metric):
    median = df[metric].median()
    iqr = df[metric].quantile(0.75) - df[metric].quantile(0.25)
    return median, iqr

# Function to perform Wilcoxon test
def wilcoxon_test(baseline, model, metric):
    stat, p_value = stats.wilcoxon(baseline[metric], model[metric])
    return stat, p_value

# Metrics to evaluate
metrics = ["test_IoUs", "test_accs", "test_eces"]
models = {"MCD": mcd_df, "OHEM": ohem_df, "EDL": edl_df}

# Compute results
for model_name, model_df in models.items():
    print(f"\n--- {model_name} vs Baseline ---")
    for metric in metrics:
        median_baseline, iqr_baseline = compute_median_iqr(baseline_df, metric)
        median_model, iqr_model = compute_median_iqr(model_df, metric)
        stat, p_value = wilcoxon_test(baseline_df, model_df, metric)

        print(f"{metric}:")
        print(f"  Baseline  → Median: {median_baseline:.4f}, IQR: {iqr_baseline:.4f}")
        print(f"  {model_name} → Median: {median_model:.4f}, IQR: {iqr_model:.4f}")
        print(f"  Wilcoxon Test: W = {stat:.2f}, p = {p_value} {'(✓ Significant)' if p_value < 0.05 else '(ns Not Significant)'}")
```
